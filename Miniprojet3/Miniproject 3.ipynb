## Plot raw data

```python
def plot_emg_signals(emg, fs, time_stop = emg.shape[0]):
    """
    This function plots EMG (Electromyography) signals.

    Parameters:
    emg (numpy.ndarray): A 2D array containing EMG data. Each column represents a different EMG sensor.
    fs (int): The sampling frequency of the EMG data.

    Returns:
    None: This function doesn't return anything but plots the EMG signals.
    """
    tsteps = np.arange(0, time_stop/fs, 1/fs)
    n_channels = emg.shape[1]

    fig, axes = plt.subplots(n_channels, 1, sharex=True, constrained_layout=True, figsize=(15, 2*n_channels))
    
    for i in range(n_channels):
        axes[i].plot(tsteps, emg[:time_stop, i])
        axes[i].set_title(f'EMG Signal from Sensor {i+1}')
        axes[i].set_xlabel("Time [s]")
        axes[i].set_ylabel("Signal [V]")
        
    plt.show()
```

## Preprocessing step : bandpass filter and powergrid noise removal 

> As first preprocessing step, we need to apply a bandpass filter. In particular, in this case we use a butterworth filter between 5Hz and 500Hz. The lower bound removes mainly movement artefacts while the upper bound removes high-frequency noise. Also, we have to remember to remove the powergrid noise (50Hz and its harmonics).

Look at https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.sosfiltfilt.html


```python
def preprocess_emg(emg, fs, bandpass_range=(5, 500), powergrid_noise_harmonics=range(50, 550, 50)):
    # Apply bandpass filter
    sos = butter(N=4, Wn=bandpass_range, fs=fs, btype="bandpass", output="sos")
    emg_filtered = sosfiltfilt(sos, emg.T).T

    # Remove powergrid noise and its harmonics
    for Wn in powergrid_noise_harmonics:
        sos = butter(N=4, Wn=(Wn - 0.5, Wn + 0.5), fs=fs, btype="bandstop", output="sos")
        emg_filtered = sosfiltfilt(sos, emg_filtered.T).T

    return emg_filtered
```

### Preprocessing step : rectified the signal to better quantify the muscular activation.

```python
def to_emg_envelope(emg_rectified, mov_mean_size = 400):
    """
    Calculate the EMG envelope of a rectified EMG signal using a moving average filter.

    The function computes the envelope of an electromyography (EMG) signal by applying a moving 
    average filter to the rectified signal. The size of the moving average window is specified 
    by the user.

    Parameters:
    emg_rectified (numpy.ndarray): The rectified EMG signal. This is typically the absolute 
                                   value of the original EMG signal.
    mov_mean_size (int): The size of the moving average window. This determines how many 
                         consecutive samples are used to compute the average at each point.

    Returns:
    numpy.ndarray: The EMG envelope signal, representing the smoothed version of the rectified 
                   EMG signal.
    """

    # Create the moving average weights
    mov_mean_weights = np.ones(mov_mean_size) / mov_mean_size

    # Calculate the EMG envelope using convolution
    emg_envelopes = convolve1d(emg_rectified, weights=mov_mean_weights, axis=0)

    return emg_envelopes
```

```python
def plot_emg_and_glove_signal(emg_envelopes, glove, time_end=emg_envelopes.shape[0], electrode=0, sensor=0):
    """
    Plots EMG envelopes and glove sensor data on the same graph with dual y-axes.

    This function takes processed EMG envelopes and glove sensor data, and plots them 
    on a single graph with two y-axes to visualize the relationship between muscle activity 
    and sensor readings.

    Parameters:
    emg_envelopes (numpy.ndarray): The EMG envelope data.
    glove (numpy.ndarray): The glove data from sensors.
    time_end (int, optional): The number of time samples to plot. Defaults to the full length of emg_envelopes.
    electrode (int, optional): The index of the electrode in the EMG data to plot. Defaults to 0.
    sensor (int, optional): The index of the sensor in the glove data to plot. Defaults to 0.

    Note:
    - The `electrode` parameter specifies which column of `emg_envelopes` to plot.
    - The `sensor` parameter specifies which column of `glove` to plot.
    """

    # Default to full length if time_end is not specified
    if time_end is None:
        time_end = emg_envelopes.shape[0]

    fig, ax = plt.subplots()
    ax.plot(emg_envelopes[:time_end, electrode], color='cornflowerblue', label="EMG envelope")
    ax.set_ylabel(f"EMG envelope (Signal {electrode})")
    #ax.set_ylim([0, 0.00017])  # For first y-axis
    ax.set_ylim([0, np.max(emg_envelopes[:time_end, electrode])*1.1]) 
    ax.set_xlabel("Time samples")
    ax.grid(False)

    ax2 = ax.twinx()
    ax2.plot(glove[:time_end, sensor], color="darkorange", label=f"Glove Signal (Sensor {sensor})")
    ax2.set_ylabel(f"Glove Signal (Sensor {sensor})")
    #ax2.set_ylim([-15, 30])  # For second y-axis
    ax2.set_ylim([np.min(glove[:time_end, sensor])*2, np.max(glove[:time_end, sensor])*1.1])
    ax2.grid(False)

    # Get handles and labels for legend
    handles1, labels1 = ax.get_legend_handles_labels()
    handles2, labels2 = ax2.get_legend_handles_labels()

    # Combine handles and labels
    handles = handles1 + handles2
    labels = labels1 + labels2

    # Create a single legend
    ax.legend(handles, labels, loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
def cut_datasets(EMG, Labels, train_cut, test_cut, val_cut):

    """
    This function is defined to cut the data in three sets
    :param EMG: Numpy array containing the data
    :param Targets: Numpy array containing the targets
    :param val_cut: information on how/where to cut the dataset to obtain the validation set
    :param test_cut: information on how/where to cut the dataset to obtain the test set
    :return: 6 pandas DataFrames (or numpy arrays) containing EMG and Labels of each sets
    """

    EMG_train = EMG[:train_cut]
    EMG_test = EMG[train_cut:test_cut]
    EMG_val = EMG[test_cut:val_cut]
   
    Labels_train = Labels[:train_cut]
    Labels_test = Labels[train_cut:test_cut]
    Labels_val = EMG[test_cut:val_cut]
    

    return EMG_train, EMG_val, EMG_test, Labels_train, Labels_val, Labels_test

```

```python
def extract_time_windows_regression(EMG,Label, fs,win_len,step):
# This function is used to cut the time windows from the raw EMG 
# It return a lists containing the EMG of each time window.
# It also returns the target corresponding to the time of the end of the window
    """
    This function is defined to perform an overlapping sliding window 
    :param EMG: Numpy array containing the data
    :param Targets: Numpy array containing the targets
    :param fs: the sampling frequency of the signal
    :param win_len: The size of the windows (in seconds)
    :param step: The step size between windows (in seconds)
    :return: A Numpy array containing the windows
    :return: A Numpy array containing the targets aligned for each window
    :note: The length of both outputs should be the same
    """
    
    n,m = EMG.shape
    win_len = int(win_len*fs)
    start_points = np.arange(0,n-win_len,int(step*fs))
    end_points = start_points + win_len

    EMG_windows = np.zeros((len(start_points),win_len,m))
    Labels_window = np.zeros((len(start_points),win_len,Label.shape[1]))
    for i in range(len(start_points)):
        EMG_windows[i,:,:] = EMG[start_points[i]:end_points[i],:]
        Labels_window[i,:,:] = Label[start_points[i]:end_points[i],:]
    

    return EMG_windows, Labels_window
```

```python
def extract_mean_std(EMG_windows,Labels_windows):
    """
    This function is defined to extract the mean and standard deviation of each window
    :param EMG_windows: A Numpy array containing the windows
    :return: A Numpy array containing the mean and standard deviation of each window and the mean of the labels window
    """
    
    EMG_mean = np.mean(EMG_windows,axis=1)
    EMG_std = np.std(EMG_windows,axis=1) 
    Labels_mean = np.mean(Labels_windows,axis=1)
    
    # Concatenate the mean and std of each window
    EMG_mean_std = np.concatenate((EMG_mean, EMG_std), axis=1)
    return EMG_mean_std, Labels_mean
```

```python
def calc_fft_power(EMG_windows, fs):
    N = EMG_windows.shape[1]  # Number of points in each window
    freqs = np.fft.rfftfreq(N, 1/fs)  # Frequency bins

    # Fast Fourier Transform (FFT)
    fft_vals = np.fft.rfft(EMG_windows, axis=1)
    fft_power = np.abs(fft_vals) ** 2  # Power spectrum
    return freqs[1:], fft_power[:,1:,:]


def extract_features(EMG_windows, fs):
    """
    This function extracts features from raw EMG data. 
    The output matrix containing the features is called X (N_windows*N_features)
    Each feature you decide to extract should be extracted for each EMG channel of each window
    :param windows: List of pandas DataFrames (or numpy arrays) containing the windows
    :return: The matrix X
    """
    # Mean absolute value (MAV)
    mav =  np.mean(abs(EMG_windows), axis=1)
    print(f"mav: {mav.shape}")

    # Maximum absolute Value (MaxAV)
    maxav = np.max(abs(EMG_windows), axis=1)
    print(f"maxav: {maxav.shape}")
    
    # Standard Deviation (STD)
    std = np.std(EMG_windows, axis=1)
    print(f"std: {std.shape}")

    # Root mean square (RMS)
    rms = np.sqrt(np.mean(EMG_windows, axis=1))
    print(f"rms: {rms.shape}")

    # Wavelength (WL)
    wl = np.sum(np.abs(np.diff(EMG_windows, axis=1)), axis=1) 
    print(f"wl: {wl.shape}")

    # Slope sign changes (SSC)
    diff = np.diff(EMG_windows, axis=1)
    #ssc = np.sum(np.diff(np.sign(diff) != 0, axis=1), axis =1)
    ssc = np.sum(np.diff(np.sign(diff), axis=1), axis=1)
    print(f"ssc: {ssc.shape}")
    
    # Get frequency and spectrogram power 
    freqs, fft_power = calc_fft_power(EMG_windows, fs=fs)

    # Mean power 
    mean_power = np.mean(fft_power, axis=1)
    print(f"mean power: {mean_power.shape}")

    # Total power
    tot_power = np.sum(fft_power, axis=1)
    print(f"total power: {tot_power.shape}")

    # Mean frequency (sum of the product of spectrogram power and frequency, divided by total sum of spectrogram power)
    freqs_reshaped = freqs.reshape(1, freqs.shape[0], 1) #reshape for multiplication of spectrogram power and frequency 
    mean_frequency = np.sum(freqs_reshaped, axis=1) / np.sum(fft_power, axis=1)
    print(f"mean frequency: {mean_frequency.shape}")

    # Median frequency 
    cumulative_power = np.cumsum(fft_power, axis=1)
    total_power = cumulative_power[:, -1, :]
    median_frequency = np.zeros((EMG_windows.shape[0],EMG_windows.shape[2]))

    for i in range(EMG_windows.shape[0]):
        for j in range(EMG_windows.shape[2]):
            median_frequency[i,j] = freqs[np.where(cumulative_power[i, :, j] >= total_power[i,j] / 2)[0][0]]
    print(f"median frequency: {median_frequency.shape}")
    
    # Peak frequency (use np.argmax)
    peak_frequency = np.argmax(fft_power, axis=1)
    print(f"peak frequency: {peak_frequency.shape}")

    X = np.column_stack((mav, maxav, std, rms, wl, ssc, mean_power, tot_power, mean_frequency, median_frequency, peak_frequency))

    return X
```
